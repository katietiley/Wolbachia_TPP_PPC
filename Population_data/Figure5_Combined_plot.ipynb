### 1. Format data
#### Panel A

```python
RRfig_df = pd.read_csv(f'{filepath}/RRfig_df_01-02-24.csv')
```

```python
RRfig_df
```

```python
a = list(RRfig_df.loc[RRfig_df['Suppression type'] == 'No Suppression', '%lit'])
a = a[0]
RRfig_df['20%'] = RRfig_df['20%']/a
RRfig_df['50%'] = RRfig_df['50%']/a
RRfig_df['80%'] = RRfig_df['80%']/a
RRfig_df['%lit'] = RRfig_df['%lit']/a
RRfig_df

#RRfig_df.to_csv('RRfig_norm_df.csv', index = False)
```

```python
s1 = RRfig_df.loc[1]
s1 = s1.values.tolist()
s2 = RRfig_df.loc[2]
s2 = s2.values.tolist()
s3 = RRfig_df.loc[3]
s3 = s3.values.tolist()
s4 = RRfig_df.loc[4]
s4 = s4.values.tolist()
s5 = RRfig_df.loc[5]
s5 = s5.values.tolist()
s6 = RRfig_df.loc[6]
s6 = s6.values.tolist()
```

#### Panel B

```python
weekfig_df = pd.read_csv(f'{filepath}/weekfig_df_01-02-24.csv')
```

```python
weekfig_df.drop('Suppression type', inplace=True, axis=1)
a = 0.0775 # RR for no suppression
weekfig_df['1wk'] = weekfig_df['1wk']/a
weekfig_df['2wk'] = weekfig_df['2wk']/a
weekfig_df['3wk'] = weekfig_df['3wk']/a
weekfig_df['4wk'] = weekfig_df['4wk']/a
weekfig_df['5wk'] = weekfig_df['5wk']/a
weekfig_df['6wk'] = weekfig_df['6wk']/a
weekfig_df['7wk'] = weekfig_df['7wk']/a
weekfig_df['8wk'] = weekfig_df['8wk']/a
weekfig_df['9wk'] = weekfig_df['9wk']/a
weekfig_df['10wk'] = weekfig_df['10wk']/a
weekfig_df
```

```python
w1 = weekfig_df.loc[0]
w1 = w1.values.tolist()
w2 = weekfig_df.loc[1]
w2 = w2.values.tolist()
w3 = weekfig_df.loc[2]
w3 = w3.values.tolist()
w4 = weekfig_df.loc[3]
w4 = w4.values.tolist()
w5 = weekfig_df.loc[4]
w5 = w5.values.tolist()
w6 = weekfig_df.loc[5]
w6 = w6.values.tolist()

t = np.linspace(1,10,10)
```
